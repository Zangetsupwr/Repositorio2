```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cargar el dataset
df = sns.load_dataset('tips')
df
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', style='smoker', s=60, ax=ax)
ax.set_title('RelaciÃ³n entre Total Bill y Tip con Anotaciones y Flechas', fontsize=20)
ax.set_xlabel('Total Bill ($)', fontsize=15)
ax.set_ylabel('Tip ($)', fontsize=15)
ax.legend(loc='upper right', fontsize='large', shadow=True, title='Detalles')

# AÃ±adir cuadrÃ­cula
ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')

# AÃ±adir anotaciÃ³n
ax.annotate('Mayor tip en la cena', xy=(df['total_bill'].max(), df['tip'].max()), xytext=(df['total_bill'].max()-10, df['tip'].max()-2),
            arrowprops=dict(facecolor='black', shrink=0.05))

# AÃ±adir flecha
# Usar annotate con arrowprops para asegurar que la flecha se muestre correctamente
ax.annotate(
    '',  # sin texto
    xy=(df['total_bill'].max(), df['tip'].max()),  # punta de la flecha
    xytext=(df['total_bill'].max()-5, df['tip'].max()-1),  # inicio de la flecha
    arrowprops=dict(facecolor='green', edgecolor='green', arrowstyle='->', lw=2)
)
ax.arrow(df['total_bill'].max()-5, df['tip'].max()-1, 4, 0.8, head_width=0.8, head_length=0.5, fc='green', ec='green')

plt.show()
```
