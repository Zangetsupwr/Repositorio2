```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path = "Clase 11 - Data.csv"
df = pd.read_csv(path)
df
```

```python
# Usando un histograma analiza cual es la distribucion de las edad de los alumnos.
plt.hist(df["Edad"], bins = 20, color = "blue", edgecolor="black")
plt.title("Distribución de las edades ", fontweight="bold")
```

```python
# Usando boxplot analiza la distribucion de las notas de los alumnos.
plt.boxplot(df["Nota_Final"], vert=False, patch_artist=True)
plt.title("Distribucion de las notas de los alumnos")
plt.xlabel("Notas Finales de los Alumnos")
plt.ylabel("Alumnos")
plt.show()
```

```python
#Usando boxplot analiza las notas de los alumnos por sexo.
df.groupby("Sexo")["Nota_Final"].mean()
plt.boxplot([df[df["Sexo"] == "M"]["Nota_Final"], df[df["Sexo"] == "F"]["Nota_Final"]])
plt.title("Notas Finales de los Alumnos por Sexo")
plt.xticks([1, 2], ["Masculino", "Femenino"])
plt.xlabel("Sexo")
plt.ylabel("Notas Finales")
plt.show()
```

```python
# Crea un histograma de Horas_Estudio filtrando solo a los estudiantes de Lima y Bogotá. 
df_filtro = df[df["Ciudad"].isin(["Lima", "Bogotá"])]
plt.subplot(1,2,1)
plt.hist(df_filtro[df_filtro["Ciudad"] == "Lima"]["Horas_Estudio"], bins=6, color="green", edgecolor="black")
plt.title("Horas de Estudio - Lima")
plt.subplot(1,2,2)
plt.hist(df_filtro[df_filtro["Ciudad"] == "Bogotá"]["Horas_Estudio"], bins=6, color="yellow", edgecolor="black")
plt.title("Horas de Estudio - Bogotá")
plt.show()

# ¿Qué ciudad tiene mayor número de estudiantes con más horas de estudio?
# Respuesta: Bogotá tiene mas horas de estduio que lima
```

```python
# Crea un boxplot de Ingreso_Familiar por Carrera. ¿Cuál carrera tiene mayor dispersión de ingresos familiares? ¿Y la menor?
carreras =df["Carrera"].unique()
list_box = []
for carrera in carreras:
    datos = df[df["Carrera"] == carrera]["Ingreso_Familiar"]
    list_box.append(datos)
plt.boxplot(list_box,labels=carreras)
plt.title("Ingreso Familiar por Carrera")
plt.xlabel("Carreras")
plt.ylabel("Ingreso Familiar")
plt.xticks(rotation=45)
plt.show()
```

```python
filtro_aprobados = df["Aprobado"] == "Sí"
filtro_reprobados = df["Aprobado"] == "No"

hora_estudio1 = df[filtro_aprobados]["Horas_Estudio"]
hora_estudio2 = df[filtro_reprobados]["Horas_Estudio"]

plt.Figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.hist(hora_estudio1,bins=10,color="blue")
plt.title("Aprobados")

plt.subplot(1,2,2)
plt.hist(hora_estudio2,bins=10,color="red")
plt.title("Reprobados")

plt.tight_layout()

plt.show()
```
