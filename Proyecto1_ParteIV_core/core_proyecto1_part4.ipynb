```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
path ="retail_sales.csv"
df = pd.read_csv(path)
df
```

```python
# EstadÃ­sticas descriptivas bÃ¡sicas
df.describe().T
```

```python
# VisualizaciÃ³n de la distribuciÃ³n de variables numÃ©ricas clave con histogramas y boxplots
numeric_cols = ['Age', 'Quantity', 'Price per Unit', 'Total Amount']

for col in numeric_cols:
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))
    df[col].hist(ax=axes[0], bins=20, color='skyblue', edgecolor='black')
    axes[0].set_title(f'Histograma de {col}')
    axes[0].set_xlabel(col)
    axes[0].set_ylabel('Frecuencia')
    
    df.boxplot(column=col, ax=axes[1])
    axes[1].set_title(f'Boxplot de {col}')
    plt.tight_layout()
    plt.show()
```

```python
# Crea grÃ¡ficos de lÃ­neas para mostrar las tendencias de ventas a lo largo del tiempo.
# Conviertir la columna 'Date' a tipo datetime si aÃºn no lo estÃ¡
df['Date'] = pd.to_datetime(df['Date'])

# Agrupar por fecha y suma el monto total de ventas por dÃ­a
ventas_por_fecha = df.groupby('Date')['Total Amount'].sum()

# Grafica la serie temporal de ventas
ventas_por_fecha.plot(kind="line", figsize=(10,5))
plt.title("Ventas totales a lo largo del tiempo")
plt.xlabel("Fechas de las Ventas")
plt.ylabel("Cantidades")
plt.show()
```

```python
#Utiliza grÃ¡ficos de dispersiÃ³n para analizar la relaciÃ³n entre diferentes variables.
# GrÃ¡fico de dispersiÃ³n entre Edad y Total Amount
plt.figure(figsize=(7,5))
plt.scatter(df['Age'], df['Total Amount'], alpha=0.5)
plt.title('RelaciÃ³n entre Edad y Monto Total de Compra')
plt.xlabel('Edad')
plt.ylabel('Monto Total')
plt.show()

# GrÃ¡fico de dispersiÃ³n entre Quantity y Total Amount
plt.figure(figsize=(7,5))
plt.scatter(df['Quantity'], df['Total Amount'], alpha=0.5, color='orange')
plt.title('RelaciÃ³n entre Cantidad y Monto Total de Compra')
plt.xlabel('Cantidad')
plt.ylabel('Monto Total')
plt.show()

```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Histograma de Total Amount
df['Total Amount'].hist(ax=axes[0], bins=20, color='lightgreen', edgecolor='black')
axes[0].set_title('Histograma de Total Amount')
axes[0].set_xlabel('Total Amount')
axes[0].set_ylabel('Frecuencia')

# Boxplot de Total Amount
df.boxplot(column='Total Amount', ax=axes[1])
axes[1].set_title('Grafico de caja de Total Amount')

plt.tight_layout()
plt.show()

```
