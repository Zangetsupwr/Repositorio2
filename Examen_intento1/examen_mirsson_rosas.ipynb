```python
#Crear dos tipos diferentes de visualizaciones exploratorias univariadas. Cada visualización debe incluir una breve interpretación dentro del archivo de código.
# Grafico 1: Histograma de los cargos mensuales
plt.figure(figsize=(8, 4))
df["MonthlyCharges"].hist(bins=30, color='skyblue', edgecolor='black')
plt.title("Distribución de los Cargos Mensuales")
plt.xlabel("Cargos Mensuales")
plt.ylabel("Cantidad de Clientes")
plt.show()
# *********************** Análsis del gráfico: ************************************************
# La mayoría de los clientes pagan cargos mensuales entre 20 y 80 dolares al mes. 
# Hay una ligera concentración hacia cargos más bajos, pero también existen clientes con cargos altos.
#**********************************************************************************

# Grafico 2: Gráfico de barras de la variable categórica "Churn"
plt.figure(figsize=(6, 4))
df["Churn"].value_counts().plot(kind='bar', color=['lightgreen', 'salmon'])
plt.title("Distribución de Cancelación de Clientes (Churn)")
plt.xlabel("Cancelacion de suscripciones")
plt.ylabel("Cantidad de Clientes")
plt.xticks(rotation=0)
plt.show()

# *********************** Análsis del gráfico numero 2: ************************************************
# La mayoria de los clientes frecuentes cancelan su suscripcion por motivos de no cumplir con los requisitos de suscripción.
# se observa que pocos clientes pagan su suscripcion .
# 
```

```python
#Crear dos tipos diferentes de visualizaciones exploratorias multivariadas. Cada visualización debe incluir una breve interpretación dentro del archivo de código.

# Gráfico de barras de multiples lineas segmentada por cancelación en los clientes
plt.figure(figsize=(8, 5))
pd.crosstab(df['MultipleLines'], df['Churn']).plot(
    kind='bar',
    stacked=True,
    color=['lightgreen', 'salmon'],
    edgecolor='black'
)
plt.title("Distribución de Multiples lineas por Cancelación")
plt.xlabel("Multiples Lineas ")
plt.ylabel("Cantidad de Clientes")
plt.xticks(rotation=0)
plt.show()
# A partir de estos datos datos en el grafico se puede inferir que la mayoria
# de los clientes que cancelaron su contrato , sin embargo, la mayoria de los clientes que no cancelaron su contrato
# tienen múltiples lineas contratadas.

# Gráfico de queso para la proporción de clientes por tipo de contrato
plt.figure(figsize=(6, 6))
df['Contract'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    colors=['#66b3ff', '#99ff99', '#ffcc99'],
    wedgeprops={'edgecolor': 'black'}
)
plt.title("Proporción de Clientes por Tipo de Contrato")
plt.legend()
plt.show()
# *********************** Análisis del gráfico de queso: ************************************************
# Se observa la distribución de los clientes según el tipo de contrato. 
# La mayoría de los clientes tienen contratos mes a mes, mientras que una menor proporción tiene contratos de uno o dos años.



```
