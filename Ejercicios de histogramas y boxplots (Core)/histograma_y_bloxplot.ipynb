```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "2015.csv"
data_frame = pd.read_csv(path)
data_frame
```

```python
# Crea histogramas para las variables Score, GDP per capita, Healthy life expectancy y Freedom to make life choices.
data_frame['Happiness Score'].hist(bins=20, alpha=0.7, color='green', edgecolor='black')
plt.title('DistribuciÃ³n de Happiness Score')
plt.xlabel('Happiness Score')
plt.ylabel('Frecuencia')
plt.show()

# Analisis del grafico de puntaje de felicidad 
# Se observa  que el puntaje de felicidad es entre 4 y 6 en la mayoria de las personas

data_frame['Economy (GDP per Capita)'].hist(bins=20, alpha=0.7 , color='blue',edgecolor='black')
plt.title('DistribuciÃ³n de GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Frecuencia')
plt.show()

# Se observa que el per capita  de los paises con mayor GDP es entre 100 y 150 millones de dÃ³lares

data_frame['Health (Life Expectancy)'].hist(bins=20, alpha=0.7 , color="orange",edgecolor='black')
plt.title('DistribuciÃ³n de Healthy Life Expectancy')
plt.xlabel('Healthy Life Expectancy')
plt.ylabel('Frecuencia')
plt.show()

data_frame['Freedom'].hist(bins=20, alpha=0.7, color="lightblue",edgecolor='black')
plt.title('DistribuciÃ³n de Freedom to Make Life Choices')
plt.xlabel('Freedom to Make Life Choices')
plt.ylabel('Frecuencia')
plt.show()
```

```python
#Crea boxplots para las mismas variables para visualizar la mediana, el rango intercuartil y los valores atÃ­picos.
variables = ['Happiness Score', 'Economy (GDP per Capita)', 'Health (Life Expectancy)', 'Freedom']
data_frame[variables].boxplot()
plt.title('Boxplots de Happiness Score, GDP per Capita, Healthy Life Expectancy y Freedom')
plt.xlabel('Distribucion general ')
plt.ylabel('Valores')
plt.xticks(rotation=45)
plt.show()
```

```python
# Combina los histogramas y los boxplots en una sola figura para cada variable.
fig, axes = plt.subplots(len(variables), 2, figsize=(10, 16))

for i, var in enumerate(variables):
    # Histograma
    data_frame[var].hist(ax=axes[i, 0], bins=20, alpha=0.7, color='skyblue', edgecolor='black')
    axes[i, 0].set_title(f'Histograma de {var}')
    axes[i, 0].set_xlabel(var)
    axes[i, 0].set_ylabel('Frecuencia')
    
    # Boxplot
    axes[i, 1].boxplot(data_frame[var], vert=False)
    axes[i, 1].set_title(f'Boxplot de {var}')
    axes[i, 1].set_xlabel(var)

plt.tight_layout()
plt.show()
    
```

```python
# AnÃ¡lisis de las distribuciones:

# Happiness Score: La distribuciÃ³n del puntaje de felicidad muestra que la mayorÃ­a de los paÃ­ses tienen valores entre 4 y 6, con pocos paÃ­ses en los extremos. 
# El boxplot indica una mediana clara y algunos valores atÃ­picos en ambos extremos, lo que sugiere que existen paÃ­ses significativamente mÃ¡s felices o menos felices que el promedio.

# Economy (GDP per Capita): La distribuciÃ³n es asimÃ©trica hacia la derecha, con la mayorÃ­a de los paÃ­ses concentrados en valores bajos de PIB per cÃ¡pita y unos pocos con valores muy altos. 
# El boxplot revela varios valores atÃ­picos en el extremo superior, correspondientes a paÃ­ses con economÃ­as muy desarrolladas.

# Health (Life Expectancy): La mayorÃ­a de los paÃ­ses se agrupan en valores medios-altos de esperanza de vida saludable. 
# El boxplot muestra una mediana elevada y algunos valores atÃ­picos en el extremo inferior, lo que indica que pocos paÃ­ses tienen una esperanza de vida significativamente menor.

# Freedom: La variable libertad presenta una distribuciÃ³n relativamente uniforme, aunque con una ligera concentraciÃ³n en valores medios. 
# El boxplot muestra algunos valores atÃ­picos bajos, lo que sugiere que ciertos paÃ­ses tienen niveles de libertad notablemente inferiores al resto.

# Conclusion:
#  las variables presentan distribuciones variadas y la presencia de valores atÃ­picos indica diferencias significativas entre paÃ­ses en cada aspecto analizado.
```
