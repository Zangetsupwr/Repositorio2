```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
path = "retail_sales.csv"
df = pd.read_csv(path)
df.head(10)
```

```python
#Realiza un anlisis de correlacin y crea un mapa de calor para visualizar las relaciones entre las variables.
correlacion_matrix = df.corr(method='pearson', numeric_only=True)
sns.heatmap(correlacion_matrix, annot=True, cmap='coolwarm')
plt.title('Mapa de calor de correlacin')
plt.show()
print(correlacion_matrix)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Edad vs Monto Total
sns.scatterplot(data=df, x='Age', y='Total Amount', ax=axs[0])
axs[0].set_title('Edad vs Monto Total')

# Cantidad vs Monto Total
sns.scatterplot(data=df, x='Quantity', y='Total Amount', ax=axs[1])
axs[1].set_title('Cantidad vs Monto Total')

# Precio Unitario vs Monto Total
sns.scatterplot(data=df, x='Price per Unit', y='Total Amount', ax=axs[2])
axs[2].set_title('Precio Unitario vs Monto Total')

plt.tight_layout()
plt.show()

# Aade cuadrculas, leyendas, anotaciones y flechas a tus grficos para mejorar la claridad y la informacin.

# Edad vs Monto Total
axs[0].grid(True)
axs[0].legend(['Datos'])
# Anotacin: valor mximo
max_idx_0 = df['Total Amount'].idxmax()
axs[0].annotate('Mximo', 
                xy=(df.loc[max_idx_0, 'Age'], df.loc[max_idx_0, 'Total Amount']),
                xytext=(df.loc[max_idx_0, 'Age']+2, df.loc[max_idx_0, 'Total Amount']+100),
                arrowprops=dict(facecolor='black', arrowstyle='->'),
                fontsize=9, color='red')

# Cantidad vs Monto Total
axs[1].grid(True)
axs[1].legend(['Datos'])
# Anotacin: valor mximo
max_idx_1 = df['Total Amount'].idxmax()
axs[1].annotate('Mximo', 
                xy=(df.loc[max_idx_1, 'Quantity'], df.loc[max_idx_1, 'Total Amount']),
                xytext=(df.loc[max_idx_1, 'Quantity']+1, df.loc[max_idx_1, 'Total Amount']+100),
                arrowprops=dict(facecolor='black', arrowstyle='->'),
                fontsize=9, color='red')

# Precio Unitario vs Monto Total
axs[2].grid(True)
axs[2].legend(['Datos'])
# Anotacin: valor mximo
max_idx_2 = df['Total Amount'].idxmax()
axs[2].annotate('Mximo', 
                xy=(df.loc[max_idx_2, 'Price per Unit'], df.loc[max_idx_2, 'Total Amount']),
                xytext=(df.loc[max_idx_2, 'Price per Unit']+5, df.loc[max_idx_2, 'Total Amount']+100),
                arrowprops=dict(facecolor='black', arrowstyle='->'),
                fontsize=9, color='red')
```
