```python
# Utilizando loc[], selecciona y muestra las filas de la 5 a la 10 (inclusive) y las columnas Producto y Tienda.
df_loc = df_1.loc[5:10,["Product Category","Customer ID"]]
df_loc
```

```python
#Basándonos en los datos existentes, crea nuevas columnas que sean útiles para el análisis. Por ejemplo, calcula el ingreso total por venta y normaliza las ventas.
df_1["Total Revenue"] = df_1["Total Amount"] * df_1["Price per Unit"]
df_1
```

```python
# Crea una columna que clasifique las ventas en categorías significativas (e.g., ‘Alta’, ‘Media’, ‘Baja’).
def clasificacion(x):
    if x > 100:
        return 'Alta'
    elif x > 50:
        return 'Media'
    else:
        return 'Baja'

df_1["Clasificación_Ventas"] = df_1["Total Amount"].apply(clasificacion)
df_1
```

```python
# Realiza agrupaciones por categorías como producto y tienda, producto y mes, etc.
df_1 = df_1.groupby(["Product Category", "Customer ID"]).agg({
    "Total Amount": "sum",
    "Price per Unit": "mean",
    "Total Revenue": "sum"
}).reset_index()

df_1
```

```python
#Aplicar funciones de agregación: Utiliza funciones como sum, mean, count, min, max, std, y var para obtener estadísticas descriptivas de cada grupo.
df_agg = df_1.groupby("Product Category").agg({
    "Total Amount": ["sum", "mean", "count", "min", "max", "std", "var"],
    "Price per Unit": ["mean"],
    "Total Revenue": ["sum"]
}).reset_index()

df_agg.round(2)
```

```python
# Calcula la desviación de cada venta respecto a la media de su grupo.
df_1["Desviación"] = df_1.groupby("Product Category")["Total Amount"].transform(lambda x: x - x.mean())
df_1
```
