```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
path = "Clase 9 - Temperaturas_Clean.csv"
df =pd.read_csv(path)
df
```

```python
# El evolutivo de la temperatura media de Madrid haciendo un grafico de lineas
df_madrid = df[df["ciudad"] == "Madrid"]
plt.plot(df_madrid["mes"],df_madrid["temp._media_(°c)"], label="Madrid", color="blue")
plt.xlabel("Mes")
plt.ylabel("Temperatura Media (°C)")
plt.title("Evolutivo de la Temperatura Media en Madrid")
plt.xticks(rotation=45) 
plt.show()
```

```python
# El evolutivo de las temperaturas minima , media y maxima de lima
df_lima = df[df["ciudad"] == "Lima"]

```

```python
# Una comparacion del evolutivo de las temperaturas medias de las diferentes ciudades de China.
df_china = df[df["pais"] =="China"]
plt.bar(df_china["ciudad"], df_china["temp._media_(°c)"])
plt.xticks(rotation=45)
plt.title("Evolución de las temperaturas medias de las diferentes ciudades de China")
plt.xlabel("Ciudades de china ")
plt.ylabel("Temperatura media (°C)")
plt.show()
```

```python
# Grafico de dispersion de las temperaturas medias de Noruega.
#Filtrado de las temperaturas de noruega
df_norway = df[df["pais"] == "Noruega"]
df_norway
# hacer el grafico de dispersion
plt.scatter(df_norway["mes"], df_norway["temp._media_(°c)"], c="orange")
plt.title("Temperaturas medias en Noruega")
plt.xlabel("Mes")
plt.ylabel("Temperatura media (°C)")
plt.xticks(rotation=45)
plt.show()
```

```python
#Grafico de dispersion comparativo de las temperaturas minimas, medias y maximas de Noruega.
df_norway = df[df["pais"] == "Noruega"]
plt.scatter(df_norway["mes"],df_norway["temp._mín_(°c)"], label="Temperatura Mínima", color="blue")
plt.scatter(df_norway["mes"],df_norway["temp._media_(°c)"], label="Temperatura Media", color="green")
plt.scatter(df_norway["mes"],df_norway["temp._máx_(°c)"], label="Temperatura Máxima", color="red")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
